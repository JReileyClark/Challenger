```python
oring_erosion_only = pd.read_csv('./o-ring-erosion-only.data',sep='\s+',header=None, names = ['oring_count','erosion','temperature','pressure','launch'])
oring_erosion_blowby = pd.read_csv('./o-ring-erosion-or-blowby.data',sep='\s+',header=None, names = ['oring_count','erosion_or_blowby','temperature','pressure','launch'])
```

```python
with pm.Model() as logistic_regression:
    #data
    temp = pm.Data("temp",oring_erosion_blowby["temperature"])
    press = pm.Data("press", oring_erosion_blowby["pressure"])
    fault = pm.Data("fault",oring_erosion_blowby["fault"])
    
    #params
    α = pm.Normal('α', 10, sigma=100)
    β_1 = pm.Normal('β_1', 0, sigma=.5)
    #β_2 = pm.Normal('β_2', 0, sd=.01)

    
    μ = pm.invlogit(α + β_1*temp)# + β_2*press)
    
    pm.Bernoulli("logit", p=μ, observed=fault)
    trace = pm.sample(draws=2000,tune=1000,chains=4)
```

```python
pm.traceplot(trace)
```

```python
pm.summary(trace)
```

```python
ppc = pm.sample_posterior_predictive(trace=trace, model=logistic_regression)
```
